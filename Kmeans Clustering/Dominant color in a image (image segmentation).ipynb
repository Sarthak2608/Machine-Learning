```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import cv2
import matplotlib.pyplot as plt
```

## In a image all pixels have three coordinate (R,G,B)
## find the k centers in the image 
## then convert the each pixel color into nearest dominant color

```python
img=cv2.imread('images/i2.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
original_shape=img.shape
```

```python
img=np.reshape(img,(-1,3))
```

```python
img.shape
```

```python
km=KMeans(n_clusters=4, random_state=0)
```

```python
km.fit(img)
```

```python
y=km.labels_
```

```python
km.cluster_centers_
```

```python
ny=y
c=np.array(km.cluster_centers_,dtype='uint8')
nimg=[]
for i in range(len(img)):
    nimg.append(c[ny[i]])
nimg=np.array(nimg)
```

```python
nimg=np.reshape(nimg,original_shape)
```

```python
plt.imshow(nimg)
plt.title("Segmented Image")
```
