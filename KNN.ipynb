```
#Knn: In this algo we find the closest k nearest point from query and then select highest frequency output
#Brute Force , Online Query ,No Training
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
def read_data(filename):
  df=pd.read_csv(filename)
  return df.values
x=read_data('/content/sample_data/xData.csv')
y=read_data('/content/sample_data/yData.csv')
```

```
print(x[0:5])
print(y[0:5])
```

```
x=x[:,1:]
y=y[:,1:]
r=x.shape[0]
print(r)
y=y[0:x.shape[0],:]
print(y.shape)
```

```

print(x.shape)
print(y.shape)
```

```
plt.scatter(x[:,0],x[:,1],c=y[:,0])
query_x=np.array([1,2])
plt.scatter(query_x[0],query_x[1],color='red')
```

```
def dis(x1,x2):
  return np.sqrt(sum((x1-x2)**2))
def KNN(X,Y,query_p,k=5):

  val=[]
  #print(X[0])
  #print(query_x)
  m=X.shape[0]
  for i in range(m):
    d=dis(query_x,X[i])
    val.append((d,Y[i]))
  #print(type(val))
  val=sorted(val)
  val=val[:k]
  val=np.array(val)
  val=np.array(val[:,1])
  new_val=np.unique(val,return_counts=True)
  max_freq_index=new_val[1].argmax()
  return new_val[0][max_freq_index]
```

```
res=KNN(x,y,[7,5])
print(res)
```
