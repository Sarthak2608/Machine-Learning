```
import cv2
from google.colab.patches import cv2_imshow
```

```
img=cv2.imread('Desert.jpg')  #open cv read the image in BGR format
print(img.shape)
```

```
cv2_imshow(img)
```

```
import matplotlib.pyplot as plt
```

```
plt.imshow(img)
```

```
rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.axis('on')
plt.imshow(gray_img)
```

```
rgb_img=img[200:800,200:600,:]
plt.imshow(rgb_img)
print(rgb_img.shape)
```

```
compress_img=cv2.resize(img,(100,100))
compress_img=cv2.cvtColor(compress_img,cv2.COLOR_BGR2RGB)
plt.imshow(compress_img)
```
